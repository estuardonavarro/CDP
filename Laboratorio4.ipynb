```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Pregunta 1:

Crear archivo .csv que contenga la información de todos los archivos

```python
lista = []
columns = None
i = 0
for filename in os.listdir('./Sales_Data'):
    df = pd.read_csv('./Sales_Data/'+filename)
    if i==0:
        columns = df.columns
    else:
        df.columns=columns
    lista.append(df)
    i+=1
data = pd.concat(lista, ignore_index=True)
data.head()
```

```python
data.to_csv('data.csv', index = False)
```

## Pregunta 2

¿Cuál es el mes con mejor venta?

```python
data.dropna(how = 'all', inplace = True)

data = data[data['Order Date']!='Order Date']
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%y %H:%M')
data['Month']= data['Order Date'].dt.month
data['Year']= data['Order Date'].dt.year
data['Quantity Ordered']= pd.to_numeric(data['Quantity Ordered'], downcast='integer')
data['Price Each']= pd.to_numeric(data['Price Each'], downcast='float')
data['Sales']= data['Quantity Ordered'] * data['Price Each']
```

```python
data.groupby(['Year','Month']).sum()
```

## Pregunta 3

¿Cuál es la ciudad que tiene mayor numero de ventas?

```python
data['City']=data['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
cities_top=data.groupby('City').sum().sort_values('Sales', ascending=False)
cities_top.head()
```
